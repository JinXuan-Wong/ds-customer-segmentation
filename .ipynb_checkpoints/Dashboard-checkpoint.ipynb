```python
# Function to filter sales data by month
def get_sales_data_for_month(month):
    sales_data = retail[retail['InvoiceMonth'] == month]
    stockcode_description = sales_data.groupby(['StockCode', 'Description'])['Quantity'].sum().reset_index()
    top_selling_items = stockcode_description.sort_values('Quantity', ascending=False).head(10)
    return top_selling_items
```

```python
def get_category_comparison_for_month(month):
    # Filter the data for the selected month
    filtered_data = retail[retail['InvoiceMonth'] == month]
    
    # Define the category columns for grouping
    category_columns = ['Category_Christmas Items', 'Category_Decorative Items', 'Category_Easter Items', 
                        'Category_Fashion Acessories', 'Category_Gardening', 'Category_Gift Items', 
                        'Category_Household Items', 'Category_Kitchenware', 'Category_Other', 
                        'Category_Stationery', 'Category_Toys']
    
    # Group by the category columns and sum the TotalPrice
    category_sales = filtered_data.groupby(category_columns)['TotalPrice'].sum().reset_index()
    
    return category_sales
```

```python
# Function to calculate annual sales for all categories
def get_annual_sales_data():
    annual_sales = retail.groupby('InvoiceMonth')['TotalPrice'].sum().reset_index()
    return annual_sales
```

```python
 from dash import html

# Function to format the cluster profiles and show only categories with value 1
def format_cluster_profiles(clusters, cluster_profiles):
    cluster_info_text = []
    for cluster in clusters:
        profile = cluster_profiles.loc[cluster].to_dict()
        
        cluster_info_text.append(html.H4(f"Cluster {cluster} Profile:"))

        # Filter out categories with value 0 and only show relevant ones
        profile_details = []
        for key, value in profile.items():
            # Filter out categories with value 0
            if value != 0:
                profile_details.append(html.Li(f"{key}: {value}"))
        
        cluster_info_text.append(html.Ul(profile_details))
    
    return cluster_info_text
```

```python
@app.callback(
    [
        Output('customer-info', 'children'),
        Output('cluster-plot', 'figure'),
        Output('category-section', 'style'),
        Output('category-dropdown', 'style'),
        Output('show-button', 'style'),
        Output('cluster-info', 'children'),
        Output('sales-data', 'children'),
        Output('monthly-sales-graph', 'figure'),
        Output('category-comparison-graph', 'figure'),
        Output('annual-sales-graph', 'figure')
    ],
    [
        Input('submit-button', 'n_clicks'),
        Input('show-button', 'n_clicks'),
        Input('month-dropdown', 'value'),
    ],
    [
        State('customer-id-input', 'value'),
        State('category-dropdown', 'value')
    ]
)
def update_info(n_clicks_submit, n_clicks_show, selected_month, customer_id, selected_category):
    ctx = callback_context

    customer_info_text = []
    fig = {}
    cluster_info_text = []
    category_section_style = {'display': 'none'}
    category_dropdown_style = {'display': 'none'}
    show_button_style = {'display': 'none'}

    # Handle the submit button logic
    if ctx.triggered and ctx.triggered[0]['prop_id'].split('.')[0] == 'submit-button':
        if not n_clicks_submit or not customer_id:
            customer_info_text = ["Please enter a valid Customer ID."]
            return customer_info_text, fig, category_section_style, category_dropdown_style, show_button_style, cluster_info_text, [], {}, {}, {}

        customer_id_str = str(customer_id).strip()
        customer_data = retail[retail['CustomerID'] == customer_id_str]
        if customer_data.empty:
            customer_info_text = [f"Customer ID {customer_id} not found."]
            return customer_info_text, fig, category_section_style, category_dropdown_style, show_button_style, cluster_info_text, [], {}, {}, {}

        clusters = customer_data['Cluster'].unique()
        sorted_clusters = sorted(clusters)

        customer_info_text = [html.H3(f"Customer {customer_id} belongs to the following clusters: {', '.join(map(str, sorted_clusters))}")]
        cluster_info_text = format_cluster_profiles(sorted_clusters, cluster_profiles)  # Ensure this returns a list

        fig = px.scatter(
            customer_data, 
            x='CustomerID', 
            y='Cluster', 
            color='Cluster',
            title=f"Cluster Visualization for Customer {customer_id}",
            hover_data=['CustomerID']
        )

        category_section_style = {'display': 'block'}
        category_dropdown_style = {'display': 'block'}
        show_button_style = {'display': 'block'}
        
        sales_data_outputs = update_sales_data(selected_month)
        return (
            customer_info_text, 
            fig, 
            category_section_style, 
            category_dropdown_style, 
            show_button_style, 
            cluster_info_text,
            *sales_data_outputs  # Unpacking sales data outputs
        )

    # Handle the show button logic
    elif ctx.triggered and ctx.triggered[0]['prop_id'].split('.')[0] == 'show-button':
        customer_id_str = str(customer_id).strip()
        customer_data = retail[retail['CustomerID'] == customer_id_str]
        category_cluster_info_text = []

        if selected_category in retail.columns:
            clusters = customer_data[customer_data[selected_category] == 1]['Cluster'].unique()
            if clusters.size == 0:
                category_cluster_info_text = [f"No clusters found for Customer ID {customer_id} in category {selected_category}."]
            else:
                category_cluster_info_text = [html.H3(f"Customer {customer_id} belongs to the following clusters in category '{selected_category}': {', '.join(map(str, clusters))}")]
                category_cluster_info_text += format_cluster_profiles(clusters, cluster_profiles)

                filtered_data = customer_data[customer_data['Cluster'].isin(clusters)]
                if not filtered_data.empty:
                    fig = px.scatter(
                        filtered_data,
                        x='CustomerID',
                        y='Cluster',
                        color='Cluster',
                        title=f"Cluster Visualization for Customer {customer_id} in category '{selected_category}'",
                        hover_data=['CustomerID']
                    )
                else:
                    category_cluster_info_text.append("No data available for the selected category.")
        else:
            category_cluster_info_text = [f"Category '{selected_category}' does not exist."]

        cluster_info_text += category_cluster_info_text  # List concatenation

        sales_data_outputs = update_sales_data(selected_month)
        sales_data_exists = len(sales_data_outputs) > 0  # Example condition

        sales_data_style = {'display': 'block'} if sales_data_exists else {'display': 'none'}
        
        return (
            customer_info_text, 
            fig, 
            category_section_style, 
            category_dropdown_style, 
            show_button_style, 
            cluster_info_text,
            sales_data_outputs[0],
            sales_data_outputs[1],  # Monthly sales figure
            sales_data_outputs[2],  # Category comparison figure
            sales_data_outputs[3]   # Annual sales figure
        )

    sales_data_outputs = update_sales_data(selected_month)
    return (
        customer_info_text, 
        fig, 
        category_section_style, 
        category_dropdown_style, 
        show_button_style, 
        cluster_info_text,
        *sales_data_outputs
    )

def update_sales_data(selected_month):
    if not selected_month:
        return [], {}, {}, {}  

    # Fetch the sales data for the selected month
    sales_data = get_sales_data_for_month(selected_month)
    category_comparison = get_category_comparison_for_month(selected_month)
    annual_sales = get_annual_sales_data()

    monthly_sales_fig = px.bar(
        sales_data,
        x='StockCode',
        y='Quantity',
        color='Description',
        title=f'Top Selling Items in Month {selected_month}'
    )

    # Create a long-form DataFrame for category comparison
    category_comparison_long = category_comparison.melt(
        id_vars=['TotalPrice'], 
        var_name='Category', 
        value_name='Sold'
    )
    
    # Filter to include only sold items
    category_comparison_long = category_comparison_long[category_comparison_long['Sold']]

    category_comparison_fig = px.bar(
        category_comparison_long,
        x='Category',
        y='TotalPrice',
        title=f'Sales Comparison Across Categories in Month {selected_month}'
    )

    annual_sales_fig = px.line(
        annual_sales,
        x='InvoiceMonth',
        y='TotalPrice',
        title='Annual Sales for All Categories'
    )

    # Format sales data as a list of dictionaries (records)
    sales_data_output = sales_data.to_dict('records')  
    
    sales_data_display = []
    for item in sales_data_output:
        sales_data_display.append(
            html.Div([
                html.P(f"StockCode: {item['StockCode']}"),
                html.P(f"Description: {item['Description']}"),
                html.P(f"Quantity: {item['Quantity']}")
            ], style={'margin-bottom': '10px'})
        )

    return (
        sales_data_display,  
        monthly_sales_fig,
        category_comparison_fig,
        annual_sales_fig
    )
```

```python
# Callback for Cluster Comparison
@app.callback(
    Output('cluster-comparison-graph', 'figure'),
    Input('compare-clusters-dropdown', 'value')  
)
def update_cluster_comparison(selected_clusters):               
    if selected_clusters and len(selected_clusters) >= 2:
        # Extract the selected clusters' data
        cluster_data = retail[retail['Cluster'].isin(selected_clusters)]
        
        cluster_summary = cluster_data.groupby('Cluster').agg({
            'TotalPrice': 'sum',  # Total spending for each cluster
            'UnitPrice': 'mean',  # Average UnitPrice for comparison
            'Category_Christmas Items': 'sum',
            'Category_Decorative Items': 'sum',
            'Category_Easter Items': 'sum',
            'Category_Fashion Acessories': 'sum',
            'Category_Gardening': 'sum',
            'Category_Gift Items': 'sum',
            'Category_Household Items': 'sum',
            'Category_Kitchenware': 'sum',
            'Category_Other': 'sum',
            'Category_Stationery': 'sum',
            'Category_Toys': 'sum'
        }).reset_index()

        fig = go.Figure(data=[
            go.Bar(x=cluster_summary['Cluster'], y=cluster_summary['TotalPrice'])
        ])
        fig.update_layout(title='Cluster Comparison of Total Spending',
                          xaxis_title='Cluster',
                          yaxis_title='Total Spending')
        
        return fig
    else:
        return go.Figure() 
```

```python
@app.callback(
    Output('monthly-sales-trend-graph', 'figure'),
    Input('monthly-cluster-dropdown', 'value'),  
)
def update_monthly_sales_trend(selected_cluster):
    if selected_cluster is not None:
        monthly_trend = retail[retail['Cluster'] == selected_cluster]
        
        monthly_trend = monthly_trend.groupby('InvoiceMonth')['TotalPrice'].sum().reset_index()
        
        fig = px.line(monthly_trend, x='InvoiceMonth', y='TotalPrice',
                      title=f'Monthly Sales Trend for Cluster {selected_cluster}')
        
        return fig
    return go.Figure()
```

```python
@app.callback(
    [
        Output('customer-agg-info', 'children'),
        Output('customer-agg-graph', 'figure'),
    ],
    [
        Input('get-customer-info-button', 'n_clicks'),
    ],
    [
        State('customer-agg-input', 'value'),
    ]
)
def update_customer_agg_info(n_clicks, customer_id):
    if n_clicks > 0 and customer_id:
        customer_id_str = str(customer_id).strip()
        
        # Filter customer_agg for the specified Customer ID
        customer_agg_data = customer_agg[customer_agg['CustomerID'] == customer_id_str]

        if customer_agg_data.empty:
            return [f"Customer ID {customer_id} not found."], {}

        # Extract relevant information
        total_quantity = customer_agg_data['TotalQuantity'].values[0]
        transaction_num = customer_agg_data['TransactionNum'].values[0]
        avg_purchase_amount = customer_agg_data['AvgPurchaseAmount'].values[0]
        total_spent = customer_agg_data['TotalSpent'].values[0]
        percent_returned = customer_agg_data['PercentReturned'].values[0]

        # Create a summary display
        info_text = [
            html.H4(f"Customer ID: {customer_id_str}"),
            html.P(f"Total Quantity Purchased: {total_quantity}"),
            html.P(f"Number of Transactions: {transaction_num}"),
            html.P(f"Average Purchase Amount: {avg_purchase_amount:.2f}"),
            html.P(f"Total Amount Spent: {total_spent:.2f}"),
            html.P(f"Percent Returned: {percent_returned:.2f}%"),
        ]

        agg_df = pd.DataFrame({
            'Metrics': ['Total Quantity', 'Transaction Num', 'Avg Purchase Amount', 'Total Spent', 'Percent Returned'],
            'Values': [total_quantity, transaction_num, avg_purchase_amount, total_spent, percent_returned],
        })

        agg_fig = px.bar(
            agg_df,
            x='Metrics',
            y='Values',
            title=f'Aggregated Data for Customer ID {customer_id_str}',
            labels={'Values': 'Values'},
            color='Metrics',  # Optional: color by metrics
            color_discrete_sequence=px.colors.qualitative.Set1
        )

        return info_text, agg_fig

    return [], {}  # Return empty values if no button clicks or input
```

```python
@app.callback(
    Output('top-n-customers-graph', 'figure'),
    [Input('top-n-input', 'value'),
     Input('top-n-button', 'n_clicks')]
)
def update_top_n_graph(n, n_clicks):
    if n_clicks == 0:
        return dash.no_update
    
    top_customers = customer_agg.nlargest(n, 'TotalSpent')  
        
    fig = px.bar(
        top_customers,
        x='CustomerID',  
        y='TotalSpent',
        title=f'Top {n} Customers by Total Spending',
        labels={'CustomerID': 'Customer ID', 'TotalSpent': 'Total Spending'}
    )
    
    return fig
```

```python
@app.callback(
    Output('category-info', 'children'),
    Output('quantity-graph', 'figure'),
    Input('category-dropdown1', 'value')
)
def Category_Info(selected_category):
    # Filter DataFrame based on selected category
    filtered_df = retail[retail[selected_category] == 1]

    # Group by InvoiceMonth and calculate total quantity per month
    monthly_sum = filtered_df.groupby('InvoiceMonth')['Quantity'].sum().reset_index()

    category_info = f"Selected Category: {selected_category}, Total Quantity: {filtered_df['Quantity'].sum()}"

    fig = px.bar(monthly_sum, x='InvoiceMonth', y='Quantity', title=f'Total Quantity Sold per Month for {selected_category}')

    return category_info, fig
```

```python
@app.callback(
    Output('category-info2', 'children'),
    Output('quantity-graph2', 'figure'),
    Input('category-dropdown2', 'value')
)
def Category_Info(selected_category):
    filtered_df = retail[retail[selected_category] == 1]

    monthly_sum = filtered_df.groupby('InvoiceMonth')['TotalPrice'].sum().reset_index()

    category_info = f"Selected Category: {selected_category}, Total Sales: {filtered_df['TotalPrice'].sum():.2f}"

    fig = px.bar(monthly_sum, x='InvoiceMonth', y='TotalPrice', title=f'Total Sales Sold per Month for {selected_category}')

    return category_info, fig
```

```python
@app.callback(
    Output('sales-data-category', 'children'),
    Output('category-sales-graph', 'figure'),
    [Input('category-dropdown4', 'value')]
)
def update_sales_data_category(selected_category):
    if not selected_category:
        return [], {}  
        
    # Filter the sales data for the selected category
    category_data = retail[retail[selected_category] == 1]  

        # Group by StockCode and Description to avoid duplicate rows
    category_data_grouped = category_data.groupby(['StockCode', 'Description']).agg({
        'Quantity': 'sum',
        'TotalPrice': 'sum'
    }).reset_index()
    
    top_sales_items = category_data_grouped.nlargest(10, 'Quantity')  
    
    category_sales_fig = px.bar(
        top_sales_items,
        x='StockCode',
        y='Quantity',
        color='Description',
        title=f'Top 10 Selling Items in {selected_category}'
    )
    
    sales_data_output = top_sales_items.to_dict('records')  
    
    sales_data_display = []
    for item in sales_data_output:
        sales_data_display.append(
            html.Div([
                html.P(f"StockCode: {item['StockCode']}"),
                html.P(f"Description: {item['Description']}"),
                html.P(f"Quantity: {item['Quantity']}")
            ], style={'margin-bottom': '10px'})
        )
    
    return sales_data_display, category_sales_fig
```
